We have three variables and are interested in connecting them:

1. Mismatches at the population level (some statistic for this - maybe fraction of negatives/positives, or mean, or coefficient of variation) 
2. Angle between two linear functions (could be predicted based on the diversity of those functions if they are positive functions only) 
3. Proportion/probability of mismatches in the observations of those functions

It would be really nice to link these variables into a single equation. If we know two of them, then we can say something about the third. This connects nicely to the new direction of the observation effect paper, where we are interested in learning things from mismatches in observations. 

```python
import numpy as np
from numpy.linalg import norm  
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic
from matplotlib.colors import ListedColormap, LinearSegmentedColormap   # For colour gradients
```

```python
def is_mismatch_fun(x, function_1, function_2): 
    if np.sign(np.dot(x, function_1)) == np.sign(np.dot(x, function_2)):
        return 0 
    else: 
        return 1
    
def proportion_mismatches(a):
    ones = np.sum(a)
    return ones/len(a)

def D(function, q):
    p = function/sum(function)
    return np.power(norm(p, q), q/(1.-q))
```

```python
S = 50
num_fun = 500
num_sim = 500
q = 2

# vectors to fill in 
theta = np.zeros(num_fun)
prediction = np.zeros(num_fun)
prediction_div = np.zeros(num_fun)
Predicted_cos_theta = np.zeros(num_fun)
Cos_theta = np.zeros(num_fun)
realised = np.zeros(num_fun)
bias = np.zeros(num_fun)
    
for n in range(num_fun):
    
    # generate functions
    narrowness_1 = np.random.uniform(0, 1.5)
    narrowness_2 = np.random.uniform(0, 1.5)
    function_1 = np.exp(np.random.normal(0, narrowness_1, size=S))                      # log normal distribution
    function_2 = np.exp(np.random.normal(0, narrowness_2, size=S)) 
    
    # measure (or predict) angle between functions
    cos_theta = np.dot(function_1, function_2) / (norm(function_1) * norm(function_2))
    Cos_theta[n] = cos_theta                                                            # renamed for loop
    theta[n] = np.arccos(cos_theta)                                                     # measured in radians
    predicted_cos_theta = (np.sqrt(D(function_1, q) * D(function_2, q)))/S              # Cos theta predicted using diversity of functions
    Predicted_cos_theta[n] = predicted_cos_theta                                        # renamed for loop
    
    # prediction of mismatches based on actual (or predicted) angle between function
    prediction[n] = np.arccos(cos_theta)/np.pi                                          
    prediction_div[n] = np.arccos(predicted_cos_theta)/np.pi                            
    
    # vectors to fill in 
    frac = np.zeros(num_sim)
    mismatch = np.zeros(num_sim)
    
    # to create a random bias at the population level (same for all simulations in second loop)
    shift = np.random.uniform(-0.3, 0.3)

    for m in range(num_sim): 

        x = np.random.normal(0, 1, S) + shift
        frac[m] = len(np.where(x<0)[0])/S
        mismatch[m] = is_mismatch_fun(x, function_1, function_2)
    
    realised[n] = len(np.where(mismatch == 1)[0])/num_sim
    bias[n] = abs(0.5 - np.mean(frac))
```

```python
grad = [(0, 0, 0), (1, 1, 1)]
cmap = LinearSegmentedColormap.from_list("cmap", grad)

plt.figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')

X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 0.5)
plt.scatter(prediction_div, realised, c = bias, s = 10, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="Difference from 0.5");
```

```python
plt.figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')

X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 0.5)
plt.scatter(Predicted_cos_theta, Cos_theta, s = 10)
plt.xlabel("predicted_cos_theta")
plt.ylabel("Cos_theta");
```

## Diversity 

```python
num_fun = 1000
num_sim = 500

# vectors to fill in 
Dimensions = np.zeros(num_fun)
Unevenness = np.zeros(num_fun)
q_diff = np.zeros(num_fun)
Cos_theta = np.zeros(num_fun)
theta = np.zeros(num_fun)
prediction = np.zeros(num_fun)
realised = np.zeros(num_fun)
bias = np.zeros(num_fun) 

for n in range(num_fun):
    
    # generate community 
    S = round(np.random.uniform(10, 100))
    Dimensions[n] = S
    unevenness = np.random.uniform(0, 10)
    Unevenness[n] = unevenness
    B = np.exp(np.random.normal(0, unevenness, size=S))
    P = B/np.sum(B)
    
    # generate diversity indices and their gradients 
    q1 = 0.01 #np.random.uniform(0.01, 3)
    q2 = 2    #np.random.uniform(0.01, 3)
    d_grad_1 = np.power(P, q1-1) - np.sum(np.power(P, q1))
    d_grad_2 = np.power(P, q2-1) - np.sum(np.power(P, q2))
    q_diff[n] = abs(q1-q2)

    # measure angle between diversity indices gradients
    cos_theta = np.dot(d_grad_1, d_grad_2) / (norm(d_grad_1) * norm(d_grad_2))  # angle between diversity gradients
    Cos_theta[n] = cos_theta                                                    # renamed for loop
    theta[n] = np.arccos(cos_theta)                                             # measured in radians
    
    # prediction of mismatches based on actual (or predicted) angle between diversity indices gradients
    prediction[n] = np.arccos(cos_theta)/np.pi 
    
    # vectors to fill in 
    frac = np.zeros(num_sim)
    mismatch = np.zeros(num_sim)
    
    # to create a random bias at the population level (same for all simulations in second loop)
    shift = np.random.uniform(-0.3, 0.3)

    for m in range(num_sim): 

        x = np.random.normal(0, 1, S) + shift
        frac[m] = len(np.where(x<0)[0])/S
        mismatch[m] = is_mismatch_fun(x, d_grad_1, d_grad_2)
    
    realised[n] = len(np.where(mismatch == 1)[0])/num_sim
    bias[n] = abs(0.5 - np.mean(frac))

```

```python
plt.hist(Cos_theta, bins = 20);
```

```python
grad = [(0, 0, 0), (1, 1, 1)]
cmap = LinearSegmentedColormap.from_list("cmap", grad)

plt.figure(num=None, figsize=(12, 12), dpi=100, facecolor='w', edgecolor='k')

plt.subplot(2, 2, 1)
X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 1)
plt.scatter(prediction, realised, c = bias, s = 5, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="Difference from 0.5");

plt.subplot(2, 2, 2)
X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 1)
plt.scatter(prediction, realised, c = Unevenness, s = 5, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="Unevenness");

plt.subplot(2, 2, 3)
X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 1)
plt.scatter(prediction, realised, c = q_diff, s = 5, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="q_diff");

plt.subplot(2, 2, 4)
X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 1)
plt.scatter(prediction, realised, c = Dimensions, s = 5, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="Dimensions");
```

```python
S = 50
num_sim = 10000

function_1 = np.random.uniform(0, 1, size=S)         
function_2 = np.random.uniform(0, 1, size=S)
cos_theta = np.dot(function_1, function_2) / (norm(function_1) * norm(function_2))
theta = np.arccos(cos_theta)
prediction = theta/np.pi

# vectors to fill in 
frac = np.zeros(num_sim)
mismatch = np.zeros(num_sim)

for n in range(num_sim): 
    x = np.random.normal(0, 1, S) #+ np.random.uniform(-0.5, 0.5)
    frac[n] = len(np.where(x<0)[0])/S
    mismatch[n] = is_mismatch(x, function_1, function_2)
```

```python
plt.figure(num=None, figsize=(12, 6), dpi=200, facecolor='w', edgecolor='k')

plt.subplot(1, 2, 1)
plt.scatter(frac, mismatch, alpha = 0.01)
plt.xlabel("Fraction of negatives")
plt.ylabel("Mismatch (1=yes)")

plt.subplot(1, 2, 2)
bin_means, bin_edges, binnumber = binned_statistic(frac, mismatch, statistic=proportion_mismatches, bins=50)
bin_width = (bin_edges[1] - bin_edges[0])
bin_2_x = bin_edges[1:] - bin_width/2
plt.scatter(bin_2_x, bin_means, alpha = 1, s =15)
plt.xlabel("Fraction of negatives")
plt.ylabel("Proportion of mismatches")
plt.axhline(y=prediction, color='r', linestyle='--');
```

```python
def hill(biomass, hill_n):
    q = hill_n
    positives = np.where(biomass > 1e-10)
    p = biomass[positives]/sum(biomass[positives])
    return np.power(norm(p, q), q/(1.-q))

biomass = np.array([60, 60])

hill(biomass, 2)
```

```python
biomass = np.array([30, 30])

hill(biomass, 2)
```

```python
biomass = np.array([55, 35])

hill(biomass, 2)
```

```python
biomass = np.array([25, 5])

hill(biomass, 2)
```

```python
x = np.linspace(1, 100, 100)
y = np.linspace(1, 100, 100)
X, Y = np.meshgrid(x, y)
Y = np.flip(Y)
joint = np.stack((X,Y),axis=2)
```

```python
def rows(c, q):
    row = np.zeros(len(x))
    for r in x:
        r = round(r)
        i = r-1
        j = c-1
        row[i] = D(joint[i, j], q)
    return row
```

```python
def heatmap(q): 
    table = np.zeros([len(x), len(x)])
    for c in x: 
        c = round(c)
        i = c-1
        table[:, i] = rows(i+1, q)
    return table
```

```python
div_low = heatmap(0.5)
div_mid = heatmap(1.01)
div_high = heatmap(2.0)
```

```python
grad = [(0.1, 0.1, 0.1), (1, 1, 1)]
cmap = LinearSegmentedColormap.from_list("cmap", grad)

locations = [(5, 90),
             (10, 90), 
             (20, 90), 
             (35, 90),
             (50, 90),
             (90, 50),
             (90, 35), 
             (90, 20), 
             (90, 10),
             (90, 5)]

plt.figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')

image = plt.imshow(div_high, cmap=cmap, extent=(1, 100, 1, 100), interpolation='bilinear', vmin=1, vmax=2) 
plt.colorbar()
plt.xlabel("Species 1", fontsize=15)
plt.ylabel("Species 2", fontsize=15)
plt.title("q = 2")
con = plt.contour(X, Y, div_high, colors='black', linestyles='dashed', alpha = 0.6, levels = [1.1, 1.3, 1.5, 1.7, 1.9])
plt.clabel(con, inline=True, fontsize=15, fmt = '%1.1f', manual = locations)
plt.show();
```
