```python
# General use
import numpy as np
from numpy.linalg import eig, norm, inv                                
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import odeint                                      # Solves a system of ODEs
from matplotlib.colors import ListedColormap, LinearSegmentedColormap   # For colour gradients
```

```python
def hill(biomass, hill_n):
    q = hill_n
    positives = np.where(biomass > 1e-10)
    p = biomass[positives]/sum(biomass[positives])
    return np.power(norm(p, q), q/(1.-q))

def is_mismatched(a,b): 
    if np.sign(a) == np.sign(b):
        return 0 
    else: 
        return 1
```

```python
# Function to run multiple-stressor in silico experiments that takes: 

#    num_stress      = The number of multiple stressor experiments to run per community
#    S               = Number of species in the species pool (for each community)
#    zeta            = Controls biomass distribution (for each community)
#    sI_d            = The direction/bias of stressors. 
#    sI_i            = The intensity of stressors.
#    Type_I_rel      = If "true", stressors are relative to rho, otherwise stressors are independent of rho

##############################################################

def stressor_experiment(num_stress, S, zeta, sI_d, sI_i, Type_I_rel):
    
    ### Generate community
    persisting = np.exp(np.random.normal(0, zeta, size=round(S/2)))
    extinct = - np.exp(np.random.normal(0, zeta, size=round(S/2)))
    rho = np.concatenate([persisting, extinct])
    Nf = np.maximum(rho, 0)
    
    ################## Loop to run stressor simulations on each community ###################
    # Defining vectors to be filled 
    syn_dist = np.zeros(num_stress)
    syn_rich = np.zeros(num_stress)
    syn_shannon = np.zeros(num_stress)
    syn_total = np.zeros(num_stress)

    ### Actual loop to run "num_stress" stressor expeirments on each community
    for n in range(num_stress):      

        ########## Generate Stressors ##########

        ##### Stressor 1 #####
        s1 = np.random.normal(sI_d, 1, size=S)      # Draw stressor 2 from a random normal distribution
        s1 = s1/norm(s1)                            # Normalise so that changing direction does not change intensity
        s1_int = np.random.uniform(0, sI_i)         # Intensity value to scale entire stressor vector by 
        
        if Type_I_rel == "true":
            s1 = s1 * s1_int * abs(rho)             # Stressors proportional to growth rates
        else: 
            s1 = s1 * s1_int * norm(rho)            # Stressors NOT proportional to growth rates (but scaled by abundance distribution)
        
        s1 = s1 * np.sqrt(S)                        # Multiply stressors by square root of S so intensity becomes a fraction of each species' rho

        ##### Stressor 2 #####
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        
        if Type_I_rel == "true":
            s2 = s2 * s2_int * abs(rho)                
        else: 
            s2 = s2 * s2_int * norm(rho)               
        
        s2 = s2 * np.sqrt(S)
  
        ########## Apply Stressors ##########       

        # Biomass after stressors
        NS_1 = np.maximum((rho - s1), 0)
        NS_2 = np.maximum((rho - s2), 0)
        NS_12 = np.maximum((rho - (s1+s2)), 0)
                
        # Change in biomass 
        DN_1 = NS_1 - Nf       
        DN_2 = NS_2 - Nf
        DN_12 = NS_12 - Nf
    
        # Community metrics - Stressor 1
        Rich_1= len(np.where(NS_1>0)[0]) - len(np.where(Nf>0)[0])             # Richness
        Diversity_1 = hill(NS_1, 1.01) - hill(Nf, 1.01)                       # Hill-Shannon
        Total_1 = sum(NS_1) - sum(Nf)                                         # Total Biomass

        # Community metrics - Stressor 2
        Rich_2= len(np.where(NS_2>0)[0]) - len(np.where(Nf>0)[0])             # Richness
        Diversity_2 = hill(NS_2, 1.01) - hill(Nf, 1.01)                       # Hill-Shannon
        Total_2 = sum(NS_2) - sum(Nf)                                         # Total Biomass

        # Community metrics - Stressors 1+2
        Rich_12= len(np.where(NS_12>0)[0]) - len(np.where(Nf>0)[0])           # Richness
        Diversity_12 = hill(NS_12, 1.01) - hill(Nf, 1.01)                     # Hill-Shannon
        Total_12 = sum(NS_12) - sum(Nf)                                       # Total Biomass
 
        
        ##### Vectors filled after each experiment #####   
             
        ##########   Agregate Properties   ##########
            
        ## Eucledian Distance (default of norm) ##
        syn_dist[n] = norm(DN_12) - norm((DN_1 + DN_2))      
        
        ## Richness - Absolute Synergism ##
        syn_rich[n] = abs(Rich_12) - abs((Rich_1 + Rich_2)) 
        
        ## Diversity - Absolute Synergism ## 
        syn_shannon[n] = abs(Diversity_12) - abs((Diversity_1 + Diversity_2))   
          
        ## Total Biomass - Absolute Synergism ## 
        syn_total[n] = abs(Total_12) - abs((Total_1 + Total_2))
        
        ##### Scaling up approach (will be the same for distance and total biomass and other linear functions?)
        #DN_pred = DN_1 + DN_2 
        #NS_pred = Nf + DN_pred 
        #Rich_pred = len(np.where(NS_pred>0)[0]) - len(np.where(Nf>0)[0]) 
        #Diversity_pred = hill(NS_pred, 1.01) - hill(Nf, 1.01) 
        #Total_pred = sum(NS_pred) - sum(Nf)
        #syn_total[n] = abs(Total_12) - abs(Total_pred)
        #syn_shannon[n] = abs(Diversity_12) - abs(Diversity_pred)  
        #syn_rich[n] = abs(Rich_12) - abs(Rich_pred)
        
        if n%100==0:
            print(n)
    
    print("")
    
    return syn_dist, syn_rich, syn_shannon, syn_total
```

```python
##### Functions for to create figures 

# colours
j_blue = (68/255, 114/255, 255/255)
j_yell = (255/255, 196/255, 0/255)
j_yell2 = (255/255, 150/255, 0/255)

# colour gradients
yellow_grad = ["white", j_yell]
yellow_cmap = LinearSegmentedColormap.from_list("mycmap", yellow_grad)
blue_grad = ["white", j_blue]
blue_cmap = LinearSegmentedColormap.from_list("mycmap", blue_grad)
red_grad = ["white", "red"]
red_cmap = LinearSegmentedColormap.from_list("mycmap", red_grad)

##############################################################

# Function to make the observation effect plots

#    X            = Vector of absolute synergism measured in one community metric (usually Euclidean Distance)
#    X            = Vector of absolute synergism measured in one community metric (e.g. Richess, Diversity)
#    percentile   = Include values up to this point (setting to e.g. 95 will improve clarity of plots)
#    r            = Number of rows in figure
#    c            = Number of columns in figure
#    n            = Subplot number


##############################################################


def observation_effect_plot(X, Y, percentile, r, c, n):    
    
    # subplot and setting axes (required for consistent text placement)
    ax = plt.subplot(r, c, n) 
    
    # option to filter out the outliers for clarity
    dataframe = pd.DataFrame({'X':X, 'Y':Y})
    dataframe = dataframe[dataframe.X < np.percentile(abs(dataframe.X),percentile)]
    dataframe = dataframe[dataframe.X > -np.percentile(abs(dataframe.X),percentile)]
    dataframe = dataframe[dataframe.Y < np.percentile(abs(dataframe.Y),percentile)]
    dataframe = dataframe[dataframe.Y > -np.percentile(abs(dataframe.Y),percentile)]
    X = dataframe.X.values
    Y = dataframe.Y.values

    # setting limits of the plot
    nb = sum(~np.isnan(Y))
    Max_X = max(abs(X)) + max(abs(X))*0.1
    Max_Y = max(abs(Y)) + max(abs(Y))*0.1
    plt.xlim(-Max_X, Max_X)
    plt.ylim(-Max_Y, Max_Y)

    # drawing background gradients in each quadrant 
    plt.imshow([[1.,1.], [0.,0.]],
               cmap = blue_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
              alpha = 0.5,
              extent = (0, Max_X, 0, Max_Y))   
    plt.imshow([[0.,1.], [0.,1.]],
               cmap = blue_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
              alpha = 0.25,
              extent = (0, Max_X, 0, Max_Y))      
    plt.imshow([[0.,1.], [0.,1.]],
               cmap = blue_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.5,
               extent = (0, -Max_X, 0, -Max_Y))   
    plt.imshow([[1.,1.], [0.,0.]],
               cmap = blue_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.25,
               extent = (0, -Max_X, 0, -Max_Y)) 
    plt.imshow([[1.,1.], [0.,0.]],
               cmap = red_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.1,
               extent = (0, Max_X, 0, -Max_Y)) 
    plt.imshow([[0.,1.], [0.,1.]],
               cmap = red_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.05,
               extent = (0, Max_X, 0, -Max_Y)) 
    plt.imshow([[1.,1.], [0.,0.]],
               cmap = red_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.05,
               extent = (0, -Max_X, 0, Max_Y)) 
    plt.imshow([[0.,1.], [0.,1.]],
               cmap = red_cmap, 
               interpolation = 'quadric',
               aspect = 'auto',
               alpha = 0.025,
               extent = (0, -Max_X, 0, Max_Y))  

    # plotting different coloured points
    A_A = np.where((X<0) & (Y<0))[0]
    S_S = np.where((X>0) & (Y>0))[0]
    A_S = np.where((X>0) & (Y<0))[0]
    S_A = np.where((X<0) & (Y>0))[0]
    ADD = np.where((X==0) & (Y==0))[0]
    AD_N = np.where((X==0) & (Y!=0))[0]
    N_AD = np.where((X!=0) & (Y==0))[0]
    plt.scatter(X[A_A], Y[A_A], color = j_blue, alpha = 1, zorder=150, s=10)
    plt.scatter(X[S_S], Y[S_S], color = j_blue, alpha = 1, zorder=150, s=10)
    plt.scatter(X[ADD], Y[ADD], color = (0.4, 0.4, 0.4), alpha = 1, zorder=150, s=10)
    plt.scatter(X[S_A], Y[S_A], color = 'r', alpha = 1, zorder=100, s=10)
    plt.scatter(X[A_S], Y[A_S], color = 'r', alpha = 1, zorder=100, s=10)
    plt.scatter(X[AD_N], Y[AD_N], color = 'r', alpha = 1, zorder=100, s=10)
    plt.scatter(X[N_AD], Y[N_AD], color = 'r', alpha = 1, zorder=100, s=10)

    # plotting the percentages of points falling in each quadrant
    plt.text(.98, .98, str(round(len(np.where((Y > 0) & (X > 0))[0])/nb * 100, 1))+"%", fontsize=10, 
             ha='right', va='top', transform=ax.transAxes, zorder=1000)
    plt.text(0.02, 0.02, str(round(len(np.where((Y < 0) & (X < 0))[0])/nb * 100, 1))+"%", fontsize=10,
             ha='left', va='bottom', transform=ax.transAxes, zorder=1000)
    plt.text(.98, 0.02, str(round((len(np.where((Y < 0) & (X > 0))[0])+
                                  len(np.where((Y == 0) & (X > 0))[0])+
                                  len(np.where((Y < 0) & (X == 0))[0]))/nb * 100, 1))+"%", fontsize=10,
             color = 'r', ha='right', va='bottom', transform=ax.transAxes, zorder=1000)  
    plt.text(0.02, .98, str(round((len(np.where((Y > 0) & (X < 0))[0])+
                                  len(np.where((Y == 0) & (X < 0))[0])+
                                  len(np.where((Y > 0) & (X == 0))[0]))/nb * 100, 1))+"%", fontsize=10,
             color = 'r', ha='left', va='top', transform=ax.transAxes, zorder=1000)

    
```

# Functions (extinction/invasion model)

```python
num_fun = 1000
num_sim = 500

S = 100                                 
sI_i = 2.0                               

##############################################################

# vectors to fill in 
prediction = np.zeros(num_fun)
prediction_div = np.zeros(num_fun)
realised = np.zeros(num_fun)
bias = np.zeros(num_fun)
    
for n in range(num_fun):
    
    ### Generate community
    zeta = 0.5
    persisting = np.exp(np.random.normal(0, zeta, size=round(S/2)))
    extinct = - np.exp(np.random.normal(0, zeta, size=round(S/2)))
    rho = np.concatenate([persisting, extinct])
    Nf = np.maximum(rho, 0)
    
    # generate functions
    narrowness_1 = np.random.uniform(0, 1.25)
    narrowness_2 = np.random.uniform(0, 1.25)
    function_1 = np.exp(np.random.normal(0, narrowness_1, size=S))            
    function_2 = np.exp(np.random.normal(0, narrowness_2, size=S)) 
    
    # measure (or predict) angle between functions
    cos_theta = np.dot(function_1, function_2) / (norm(function_1) * norm(function_2))
    theta = np.arccos(cos_theta)                                                        
    predicted_cos_theta = (np.sqrt(hill(function_1, 2) * hill(function_2, 2)))/S              # Cos theta predicted using diversity of functions
    predicted_theta = np.arccos(predicted_cos_theta)
    
    # prediction of mismatches based on actual (or predicted) angle between function
    prediction[n] = theta/np.pi                                          
    prediction_div[n] = predicted_theta/np.pi                            
    
    # vectors to fill in during second loop
    mismatch = np.zeros(num_sim)

    sI_d = np.random.uniform(-0.5, 0.5)  
    bias[n] = abs(sI_d)
    
    # Loop to run stressor experiments for each community 
    for m in range(num_sim): 
        
        ########## Generate Stressors ##########
        # Stressor 1 
        s1 = np.random.normal(sI_d, 1, size=S)      # Draw stressor 2 from a random normal distribution
        s1 = s1/norm(s1)                            # Normalise so that changing direction does not change intensity
        s1_int = np.random.uniform(0, sI_i)         # Intensity value to scale entire stressor vector by 
        s1 = s1 * s1_int * abs(rho)                 # Stressors proportional to growth rates
        s1 = s1 * np.sqrt(S)                        # Multiply stressors by square root of S so intensity becomes a fraction of each species' rho

        # Stressor 2 
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        s2 = s2 * s2_int * abs(rho)                
        s2 = s2 * np.sqrt(S)

        ########## Apply Stressors ##########       
        NS_1 = np.maximum((rho - s1), 0)
        NS_2 = np.maximum((rho - s2), 0)
        NS_12 = np.maximum((rho - (s1+s2)), 0)
        
        ########## Changes in Functions ##########       
        # Changes in Functions for Stressor 1
        function_1_1 = abs(np.dot(function_1, NS_1) - np.dot(function_1, Nf))
        function_2_1 = abs(np.dot(function_2, NS_1) - np.dot(function_2, Nf))
        # Changes Functions for Stressor 2
        function_1_2 = abs(np.dot(function_1, NS_2) - np.dot(function_1, Nf))
        function_2_2 = abs(np.dot(function_2, NS_2) - np.dot(function_2, Nf))
        # Changes Functions for Stressors 1+2
        function_1_12 = abs(np.dot(function_1, NS_12) - np.dot(function_1, Nf))
        function_2_12 = abs(np.dot(function_2, NS_12) - np.dot(function_2, Nf))
        # Synergism in Functions 
        syn_fun_1 = function_1_12 - (function_1_1 + function_1_2)
        syn_fun_2 = function_2_12 - (function_2_1 + function_2_2)

        ########## Mismatches #########
        mismatch[m] = is_mismatched(syn_fun_1, syn_fun_2)
        
    realised[n] = np.sum(mismatch)/len(mismatch)
    
    if n%100==0:
        print(n)
```

```python
grad = [(0, 0, 0), (1, 1, 1)]
cmap = LinearSegmentedColormap.from_list("cmap", grad)

plt.figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')

X = np.sort(prediction)
plt.plot(X, X, 'k', lw=2, linestyle = '--', alpha = 0.5)
plt.scatter(prediction, realised, c = bias, s = 7.5, cmap = cmap, edgecolors='k', linewidths=0.1)
plt.xlabel("theta/pi")
plt.ylabel("Proportion of mismatches")
plt.colorbar(label="Bias (Stressors)");
```

# Functions (generic model)

```python
num_fun = 500
num_sim = 500
S = 50                                                           

# Key parameters 
sI_d = 0           # stressor direction
sI_i = 0.1         # stressor intensity
nonadd_d = 0.3     # nonadditivity bias
nonadd_i = 0.1     # nonadditivity intensity
alpha = 0          # scaling of non-add (1 or 0, yes or no)

##############################################################

# vectors to fill in 
prediction = np.zeros(num_fun)
prediction_div = np.zeros(num_fun)
realised = np.zeros(num_fun)
bias = np.zeros(num_fun)

# Loop to create many different communities and functions
for n in range(num_fun):
    
    # generate community 
    zeta = np.random.uniform(0.5, 1.5)     
    K = np.exp(np.random.normal(0, zeta, size=S))    
    Nf = np.maximum(K, 0)
    
    # generate functions
    narrowness_1 = np.random.uniform(0, 1.3)
    narrowness_2 = np.random.uniform(0, 1.3)
    function_1 = np.exp(np.random.normal(0, narrowness_1, size=S))            
    function_2 = np.exp(np.random.normal(0, narrowness_2, size=S)) 
    
    # functions modified by alpha and K
    function_1_mod = (K**alpha) * function_1
    function_2_mod = (K**alpha) * function_2
    
    # measure (or predict) angle between functions
    cos_theta = np.dot(function_1_mod, function_2_mod) / (norm(function_1_mod) * norm(function_2_mod))
    theta = np.arccos(cos_theta)                                                        
    predicted_cos_theta = (np.sqrt(hill(function_1, 2) * hill(function_2, 2)))/S              # Cos theta predicted using diversity of functions (best when non-add is independent of K)
    predicted_theta = np.arccos(predicted_cos_theta)
    
    # prediction of mismatches based on actual (or predicted) angle between function
    prediction[n] = theta/np.pi                                          
    prediction_div[n] = predicted_theta/np.pi                            
    
    # vectors to fill in during second loop
    mismatch = np.zeros(num_sim)
    frac = np.zeros(num_sim)
    
    # Generate a bias for the vector of non-additivity at the population level
    nonadd_bias = np.random.uniform(-nonadd_d, nonadd_d)  
    
    # Loop to run stressor experiments for each community 
    for m in range(num_sim): 
        
        ########## Generate Stressors ##########

        # Stressor 1 
        s1 = np.random.normal(sI_d, 1, size=S)      # Draw stressor from a random normal distribution with a certain mean direction
        s1 = s1/norm(s1)                            # Normalise so that changing direction does not change intensity
        s1_int = np.random.uniform(0, sI_i)         # Intensity value to scale entire stressor vector by 
        s1 = s1 * s1_int * K                        # Stressor proportional to growth rates
        s1 = s1 * np.sqrt(S)                        # Multiply by square root of S so intensity becomes a fraction of each species' K

        # Stressor 2 
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        s2 = s2 * s2_int * K              
        s2 = s2 * np.sqrt(S)

        # Non-additive term 
        nonadd = np.random.normal(nonadd_bias, 1, size = S) 
        nonadd = nonadd/norm(nonadd) 
        nonadd_int = np.random.uniform(0, nonadd_i) 
        nonadd = nonadd * nonadd_int * (K**alpha)     
        nonadd = nonadd * np.sqrt(S)
        
        # Combination 
        s12 = s1 + s2 + nonadd

        ########## Apply Stressors ##########       
        NS_1 = np.maximum((K - s1), 0)
        NS_2 = np.maximum((K - s2), 0)
        NS_12 = np.maximum((K - s12), 0)
        
        # Effect of stressors (at population level)
        DS_1 = NS_1 - Nf
        DS_2 = NS_2 - Nf
        DS_12 = NS_12 - Nf
        
        ########## Actual population non-additivity (after extinctions) ##########
        pop_nonadd = DS_12 - (DS_1 + DS_2)   
        frac[m] = len(np.where(pop_nonadd<0)[0])/len(pop_nonadd)    # fraction of synergism - 0.5 would be unbiased 
        
        ########## Synergism in Functions ##########       
        syn_fun_1 = np.dot(function_1, pop_nonadd)            # projecting vector of non-additivity onto function 1
        syn_fun_2 = np.dot(function_2, pop_nonadd)            # projecting vector of non-additivity onto function 2
    
        ########## Mismatches #########
        mismatch[m] = is_mismatched(syn_fun_1, syn_fun_2)     # 1 is mismatch, 0 if same sign 
    
    realised[n] = np.sum(mismatch)/len(mismatch)              # actual proportion of mismatches
    bias[n] = abs(0.5 - np.mean(frac))                        # bias towards either antagonism or synergism 
    
    if n%100==0:
        print(n)
```

Some insights from this are: 

- The proportion of mismatches in the observations of non-additivity by two linear, positive functions can be predicted using the angle between those functions divided by pi. This only holds true when the population-level non-additivity is unbiased - in other words, half the species respond synergistically and half the sepecies respond antagonistically. When there is a bias at the population level (due to whatever reason), there will be less mismatches between functions that predicted by their angles. Even a 60:40 split between antagonism and synergism will lead to an underestimation of the proportion of mismatches. This is just the effect of those cones - easily visualized in 2D. 

- The angle between the two functions can be predicted by measuring the diversity of the two functions, which is probably far more easily accesible in empirical data. 

- With two of the three variables, 1. angle between functions, 2. mismatches between functions, and 3. population bias of non-additivity, we can say something about the third variable. Therefore, mismatches between functions can give us insights into the underlying population-level non-additivity.  

- When non-additivity is relative to K, the relevant angles between functions are now related to K and how much non-add is scaled by K (i.e. alpha). 

- Stressor direction and stressor intensity (unless there are extinctions) have no effect here. If extinctions occur (negative stressors and high intensity) non-add will be very biased so the angle between functions will overestimate the actual proportion of mismatches.

- **Species richness has an effect** - it seems to be doing something to the bias. As richness increases the bias becomes much stronger.

# Diversity indices


```python
num_fun = 1000
num_sim = 500
S = 50                                                           

# Key parameters 
sI_d = 0           # stressor direction
sI_i = 0.1         # stressor intensity
nonadd_d = 0.3     # nonadditivity bias
nonadd_i = 0.1     # nonadditivity intensity
alpha = 1          # scaling of non-add (1 or 0, yes or no)

##############################################################

# vectors to fill in 
Unevenness = np.zeros(num_fun)
q_diff = np.zeros(num_fun)
prediction = np.zeros(num_fun)
realised = np.zeros(num_fun)
realised_grad = np.zeros(num_fun)
bias = np.zeros(num_fun)
      
# Loop to create many different communities and functions
for n in range(num_fun):
    
    # generate community 
    zeta = np.random.uniform(0.5, 2.5)      # unevenness
    Unevenness[n] = zeta
    K = np.exp(np.random.normal(0, zeta, size=S))    
    Nf = np.maximum(K, 0)
    
    # generate diversity indices and their gradients 
    P = K/np.sum(K)  
    q1 = np.random.uniform(0.01, 5)
    q2 = np.random.uniform(0.01, 5)
    d_grad_1 = (np.power(P, q1-1) - np.sum(np.power(P, q1))) * (q1/(1-q1))
    d_grad_2 = (np.power(P, q2-1) - np.sum(np.power(P, q2))) * (q2/(1-q2))
    q_diff[n] = abs(q1-q2)
    
    d_grad_1_mod = (K**alpha) * d_grad_1
    d_grad_2_mod = (K**alpha) * d_grad_2
    
    
    # measure angle between diversity indices gradients
    cos_theta = np.dot(d_grad_1_mod, d_grad_2_mod) / (norm(d_grad_1_mod) * norm(d_grad_2_mod))  # angle between diversity gradients
    theta = np.arccos(cos_theta)                                                # measured in radians
    
    # prediction of mismatches based on actual (or predicted) angle between diversity indices gradients
    prediction[n] = theta/np.pi 
    
    # vectors to fill in during second loop
    mismatch = np.zeros(num_sim)
    mismatch_grad = np.zeros(num_sim)
    frac = np.zeros(num_sim)
    
    # Generate a bias for the vector of non-additivity at the population level
    nonadd_bias = np.random.uniform(-nonadd_d, nonadd_d)  
    
    # Loop to run stressor experiments for each community 
    for m in range(num_sim): 
        
        ########## Generate Stressors ##########

        # Stressor 1 
        s1 = np.random.normal(sI_d, 1, size=S)      
        s1 = s1/norm(s1)                            
        s1_int = np.random.uniform(0, sI_i)        
        s1 = s1 * s1_int * K               
        s1 = s1 * np.sqrt(S)                       

        # Stressor 2 
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        s2 = s2 * s2_int * K              
        s2 = s2 * np.sqrt(S)

        # Non-additive term 
        nonadd = np.random.normal(nonadd_bias, 1, size = S) 
        nonadd = nonadd/norm(nonadd) 
        nonadd_int = np.random.uniform(0, nonadd_i) 
        nonadd = nonadd * nonadd_int * (K**alpha) 
        nonadd = nonadd * np.sqrt(S)
        
        # Combination 
        s12 = s1 + s2 + nonadd
        
        ########## Apply stressors ##########       
        NS_1 = np.maximum((K - s1), 0)
        NS_2 = np.maximum((K - s2), 0)
        NS_12 = np.maximum((K - s12), 0)
        
        DS_1 = NS_1 - Nf
        DS_2 = NS_2 - Nf
        DS_12 = NS_12 - Nf
        
        ########## Actual population non-additivity (after extinctions) ##########
        pop_nonadd = DS_12 - (DS_1 + DS_2)   
        frac[m] = len(np.where(pop_nonadd<0)[0])/len(pop_nonadd)    
        
        ########## Changes in diversity #########     
        D_q1_1 = hill(NS_1, q1) - hill(Nf, q1)
        D_q1_2 = hill(NS_2, q1) - hill(Nf, q1)
        D_q1_12 = hill(NS_12, q1) - hill(Nf, q1)
        
        D_q2_1 = hill(NS_1, q2) - hill(Nf, q2)
        D_q2_2 = hill(NS_2, q2) - hill(Nf, q2)
        D_q2_12 = hill(NS_12, q2) - hill(Nf, q2)
        
        ########## Synergism in diversity #########
        syn_D_q1 = D_q1_12 - (D_q1_1 + D_q1_2)  
        syn_D_q2 = D_q2_12 - (D_q2_1 + D_q2_2)  

        ########## Synergism in linear part of diversity (Gradient) #########
        syn_Dgrad1 = np.dot(d_grad_1, pop_nonadd)
        syn_Dgrad2 = np.dot(d_grad_2, pop_nonadd)
    
        ########## Mismatches #########
        mismatch[m] = is_mismatched(syn_D_q1, syn_D_q2)
        mismatch_grad[m] = is_mismatched(syn_Dgrad1, syn_Dgrad2)
    
    realised[n] = np.sum(mismatch)/len(mismatch)
    realised_grad[n] = np.sum(mismatch_grad)/len(mismatch_grad)
    bias[n] = abs(0.5 - np.mean(frac))
    
    if n%100==0:
        print(n)
```

Mismatches between the observations of diversity functions can be quite well explained by the angles between their gradients. The angles between these gradients increase as: 

- the difference between the q numbers of the diversity indices increases 
- the intial community is more uneven 

Interestingly, diversity gradients are actually very similar to positive functions, especially when evenness is low. When increases in the more dominant species will decrease diversity but all the others will increase diversity. On the other hand, when evness is high, increases to most species will decrease diversity. So we can just look at communities that aren't too even - e.g. start log nomral distribution from 0.5 rather than 0. 

The role of bias: becasue these functions behave similarly to positive functions, the role of bias is similar - bias means that nonadditivity is more likely to fall in the quadrants that are purely positive or purely negativem, which for postive functions, is exactly where the cone of mismatches isn't! There will be a systematic overestimation of the proportion of mismatches. 

The role of scaling non-additivity by biomass: when non-additivity is proportional to biomass, then the most non-additivity is focused on the negative components of the diversity gradients! This means that the bias will not have the same effect that it had for positive functions. So multiplying non-add by K hides the effect of bias. 

How does unevenness of the community impact the results? 
- totally even communities (zeta = ~0): I guess there should be very few mismatches.
- very even communities (zeta = 0.5): prediction is very good, but effect of bias is not visible 
- relatively uneven communities (zeta = 1): prediction is very good, effect of bias is visible
- very uneven communities (zeta = 2): prediction is quite good, but effect of bias is not visible


Inconsistency of species' responses to stressors causes community metrics to observe change in different ways

```python
num_fun = 1000
num_sim = 500
S = 50                                                           

# Key parameters 
sI_d = 0           # stressor direction
sI_i = 0.1         # stressor intensity
nonadd_d = 0.3     # nonadditivity bias
nonadd_i = 0.1     # nonadditivity intensity
alpha = 0          # scaling of non-add (1 or 0, yes or no)
q = 2              # diversity 


##############################################################

# vectors to fill in 
Unevenness = np.zeros(num_fun)
prediction = np.zeros(num_fun)
realised = np.zeros(num_fun)
realised_grad = np.zeros(num_fun)
bias = np.zeros(num_fun)
diversity = np.zeros(num_fun)

# Loop to create many different communities and functions
for n in range(num_fun):
    
    # generate community 
    zeta = np.random.uniform(0.5, 1.5)      # unevenness
    Unevenness[n] = zeta
    K = np.exp(np.random.normal(0, zeta, size=S))    
    Nf = np.maximum(K, 0)
    diversity[n] = hill(Nf, 2)             # initial diversity
    
    # generate total biomass and gradient of a diversity index
    P = K/np.sum(K)  
    total = np.ones(S) 
    d_grad = (np.power(P, q-1) - np.sum(np.power(P, q))) * (q/(1-q))
    
    total_mod = (K**alpha) * total
    d_grad_mod = (K**alpha) * d_grad

    # measure angle between diversity indices gradients
    cos_theta = np.dot(d_grad_mod, total_mod) / (norm(d_grad_mod) * norm(total_mod))    # angle between diversity and total
    theta = np.arccos(cos_theta)                                                # measured in radians
    
    # prediction of mismatches based on actual (or predicted) angle between diversity indices gradients
    prediction[n] = theta/np.pi 
    
    # vectors to fill in during second loop
    mismatch = np.zeros(num_sim)
    mismatch_grad = np.zeros(num_sim)
    frac = np.zeros(num_sim)

    # Generate a bias for the vector of non-additivity at the population level
    nonadd_bias = np.random.uniform(-nonadd_d, nonadd_d)  
    
    # Loop to run stressor experiments for each community 
    for m in range(num_sim): 
        
        ########## Generate Stressors ##########

        # Stressor 1 
        s1 = np.random.normal(sI_d, 1, size=S)      # Draw stressor 2 from a random normal distribution
        s1 = s1/norm(s1)                            # Normalise so that changing direction does not change intensity
        s1_int = np.random.uniform(0, sI_i)         # Intensity value to scale entire stressor vector by 
        s1 = s1 * s1_int * K                        # Stressors proportional to growth rates
        s1 = s1 * np.sqrt(S)                        # Multiply stressors by square root of S so intensity becomes a fraction of each species' rho

        # Stressor 2 
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        s2 = s2 * s2_int * K              
        s2 = s2 * np.sqrt(S)

        # Non-additive term 
        nonadd = np.random.normal(nonadd_bias, 1, size = S) 
        nonadd = nonadd/norm(nonadd) 
        nonadd_int = np.random.uniform(0, nonadd_i) 
        nonadd = nonadd * nonadd_int * (K**alpha)
        nonadd = nonadd * np.sqrt(S)
        
        # Combination 
        s12 = s1 + s2 + nonadd
        
        ########## Apply stressors ##########       
        NS_1 = np.maximum((K - s1), 0)
        NS_2 = np.maximum((K - s2), 0)
        NS_12 = np.maximum((K - s12), 0)
        
        DS_1 = NS_1 - Nf
        DS_2 = NS_2 - Nf
        DS_12 = NS_12 - Nf
            
        ########## Actual population non-additivity (after extinctions) ##########
        pop_nonadd = DS_12 - (DS_1 + DS_2) # this is the same as -nonadd (if there are no extinctions)
        frac[m] = len(np.where(pop_nonadd<0)[0])/len(pop_nonadd)    # fraction of negatives - 0.5 would be unbiased 
        
        ########## Diversity #########     
        D_1 = hill(NS_1, q) - hill(Nf, q)
        D_2 = hill(NS_2, q) - hill(Nf, q)
        D_12 = hill(NS_12, q) - hill(Nf, q)
        syn_D = D_12 - (D_1 + D_2)               # synergism as seen by diversity
        syn_D_grad = np.dot(d_grad, pop_nonadd)   # synergism as seen by just the linear component of diversity
        
        ########## Total Biomass ######### 
        T_1 = sum(NS_1) - sum(Nf) 
        T_2 = sum(NS_2) - sum(Nf) 
        T_12 = sum(NS_12) - sum(Nf) 
        syn_T = T_12 - (T_1 + T_2) 
        syn_T_grad = np.dot(total, pop_nonadd)
        
        ########## Mismatches #########
        mismatch[m] = is_mismatched(syn_T, syn_D)
        mismatch_grad[m] = is_mismatched(syn_T_grad, syn_D_grad)
    
    realised[n] = np.sum(mismatch)/len(mismatch)
    realised_grad[n] = np.sum(mismatch_grad)/len(mismatch_grad)
    bias[n] = abs(0.5 - np.mean(frac))
    
    if n%100==0:
        print(n)
    
```

### Scaling of non-add by K

```python
num_fun = 1000
num_sim = 500
S = 50                                                           

# Key parameters 
sI_d = 0           # stressor direction
sI_i = 0.1         # stressor intensity
nonadd_d = 0.3     # nonadditivity bias
nonadd_i = 0.1     # nonadditivity intensity
q = 2              # diversity 


##############################################################

# vectors to fill in 
Unevenness = np.zeros(num_fun)
prediction = np.zeros(num_fun)
prediction_mod = np.zeros(num_fun)
realised = np.zeros(num_fun)
realised_grad = np.zeros(num_fun)
scaling = np.zeros(num_fun)
diversity = np.zeros(num_fun)
bias = np.zeros(num_fun)

# Loop to create many different communities and functions
for n in range(num_fun):
    
    # generate community 
    zeta = np.random.uniform(0.5, 1.5)      # unevenness
    Unevenness[n] = zeta
    K = np.exp(np.random.normal(0, zeta, size=S))    
    Nf = np.maximum(K, 0)
    diversity[n] = hill(Nf, 2)
    
    # Generate a value of alpha, that controls how much non-addd is scaled by K 
    alpha = np.random.uniform(0, 1)
    
    # generate total biomass and gradient of a diversity index
    P = K/np.sum(K)  
    total = np.ones(S) 
    d_grad = (np.power(P, q-1) - np.sum(np.power(P, q))) * (q/(1-q))

    total_mod = (K**alpha) * total
    d_grad_mod = (K**alpha) * d_grad
    
    # measure angle between diversity indices gradients
    cos_theta = np.dot(d_grad, total) / (norm(d_grad) * norm(total))        # angle between diversity and total
    cos_theta_mod = np.dot(d_grad_mod, total_mod) / (norm(d_grad_mod) * norm(total_mod))    
    theta = np.arccos(cos_theta)                                                # measured in radians
    theta_mod = np.arccos(cos_theta_mod)
    
    # prediction of mismatches based on actual (or predicted) angle between diversity indices gradients
    prediction[n] = theta/np.pi 
    prediction_mod[n] = theta_mod/np.pi
    
    # vectors to fill in during second loop
    mismatch = np.zeros(num_sim)
    mismatch_grad = np.zeros(num_sim)
    frac = np.zeros(num_sim)

    # Generate a bias for the vector of non-additivity at the population level
    nonadd_bias = np.random.uniform(-nonadd_d, nonadd_d)
    

    # Loop to run stressor experiments for each community 
    for m in range(num_sim): 
        
        ########## Generate Stressors ##########

        # Stressor 1 
        s1 = np.random.normal(sI_d, 1, size=S)      # Draw stressor 2 from a random normal distribution
        s1 = s1/norm(s1)                            # Normalise so that changing direction does not change intensity
        s1_int = np.random.uniform(0, sI_i)         # Intensity value to scale entire stressor vector by 
        s1 = s1 * s1_int * abs(K)                   # Stressors proportional to growth rates
        s1 = s1 * np.sqrt(S)                        # Multiply stressors by square root of S so intensity becomes a fraction of each species' rho

        # Stressor 2 
        s2 = np.random.normal(sI_d, 1, size=S)                       
        s2 = s2/norm(s2)                            
        s2_int = np.random.uniform(0, sI_i)     
        s2 = s2 * s2_int * abs(K)                  
        s2 = s2 * np.sqrt(S)

        # Non-additive term 
        nonadd = np.random.normal(nonadd_bias, 1, size = S) 
        nonadd = nonadd/norm(nonadd) 
        nonadd_int = np.random.uniform(0, nonadd_i) 
        nonadd = nonadd * nonadd_int * (abs(K)**alpha)     
        nonadd = nonadd * np.sqrt(S)
        
        # Combination 
        s12 = s1 + s2 + nonadd
        
        ########## Apply stressors ##########       
        NS_1 = np.maximum((K - s1), 0)
        NS_2 = np.maximum((K - s2), 0)
        NS_12 = np.maximum((K - s12), 0)
        
        DS_1 = NS_1 - Nf
        DS_2 = NS_2 - Nf
        DS_12 = NS_12 - Nf
            
        ########## Actual population non-additivity (after extinctions) ##########
        pop_nonadd = DS_12 - (DS_1 + DS_2) # this is the same as -nonadd (if there are no extinctions)
        frac[m] = len(np.where(pop_nonadd<0)[0])/len(pop_nonadd)    # fraction of negatives - 0.5 would be unbiased 
        
        ########## Diversity #########     
        D_1 = hill(NS_1, q) - hill(Nf, q)
        D_2 = hill(NS_2, q) - hill(Nf, q)
        D_12 = hill(NS_12, q) - hill(Nf, q)
        syn_D = D_12 - (D_1 + D_2)               # synergism as seen by diversity
        syn_D_grad = np.dot(d_grad, pop_nonadd)   # synergism as seen by just the linear component of diversity
        
        ########## Total Biomass ######### 
        T_1 = sum(NS_1) - sum(Nf) 
        T_2 = sum(NS_2) - sum(Nf) 
        T_12 = sum(NS_12) - sum(Nf) 
        syn_T = T_12 - (T_1 + T_2) 
        syn_T_grad = np.dot(total, pop_nonadd)
        
        ########## Mismatches #########
        mismatch[m] = is_mismatched(syn_T, syn_D)
        mismatch_grad[m] = is_mismatched(syn_T_grad, syn_D_grad)
    
    realised[n] = np.sum(mismatch)/len(mismatch)
    realised_grad[n] = np.sum(mismatch_grad)/len(mismatch_grad)
    bias[n] = abs(0.5 - np.mean(frac))
    scaling[n] = alpha
    
    if n%100==0:
        print(n)
    
```

It turns out that the relationship between proportion of mismatches and the angle between functions is strongly impacted by the "scaling" of non-additivity by the biomass of species (i.e. is nonadditivity relative to biomass). 

Is this interesting? Can the proportion of mismatches between biomass and diversity tell us something about how non-additivity is spread across species - is it relative to their biomass or independent of their biomass? 

So looking at the mismatches between observations at the community level, we can learn two things about the non-additivity at the population level: 

1. Is the non-additivity biased at the population level (are there more antagonistic or synergistic species-level responses)? 

2. Is non-additivity relative to the biomass of species? 


**Case when non-additivity is relative to biomass (white points):** 
* At the left side of the plot, the angle between diversity and biomass is small, which means that the community is very uneven. Here, non-additivity is systematically falling orthongally to the functions - due to simple effects that are not related to diversity specifically. When non-additivity is falling ortaongally to the functions, it will systematically fall in the cone of mismatches, which is also orthongal to the functions when they are colinear. 
* At the right hand side of the plot, the angle between diversity and biomass is large, which means that the community is very even. Now, the prediction is performing quite well for the diversity gradient, but for the diversities themselves this is when the non-linearity is most strong and we get this funnel effect, where due to the non-linearity, mismatches can be anything. 
* What does bias do: when there is a bias, it (slightly) reduces the effect of the scaling as it means that the rare species tend to be working together, rather than canceling each other out, to outweigh the most dominant species - so the non-additivity in a way is not as uneven as the biomass. It also exagerates the effect of the non-linearity. 

**Case when non-additivity is independent of biomass (black points):** 
* Gradients behave just like two positive functions - the role of the bias is the same. 
* Diversity itself is different as it is non-linear, so we see the same weird funnel effect and less predictable behaviour.
