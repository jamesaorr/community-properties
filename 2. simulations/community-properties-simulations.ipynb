```python
######## General use #############
import numpy as np      
import matplotlib.pyplot as plt                                         
import pandas as pd                                                

####### Specific functions ########
from matplotlib.colors import ListedColormap, LinearSegmentedColormap   # For colour gradients
from numpy.linalg import norm, eig  

####### User-defined functions ########
# Hill diversity
def hill(biomass, hill_n):
    q = hill_n
    positives = np.where(biomass > 1e-10)
    p = biomass[positives]/sum(biomass[positives])
    return np.power(norm(p, q), q/(1.-q))

# Do a and b have the same sign?
def is_mismatched(a,b): 
    if np.sign(a) == np.sign(b):
        return 0 
    else: 
        return 1

####### Colours ########
j_blue = (68/255, 114/255, 255/255)
blue_grad = ["white", j_blue]
blue_cmap = LinearSegmentedColormap.from_list("mycmap", blue_grad)
red_grad = ["white", "red"]
red_cmap = LinearSegmentedColormap.from_list("mycmap", red_grad)
 
```

## Function for simulations  <a class="anchor" id="3"></a>
[Top](#1)

```python
### Function to generate many communities and perturbations and then compare observations

#    num_funs        = Number of function pairs (and communities) - points on the plots
#    num_exp         = Number of experiments for each function pair - simulations within a point
#    P_d             = Average perturbation direction
#    P_i             = Upper limit of perturbation intensity
#    alpha_low       = Scaling of perturbation by biomass between 0 and 1 (lower bound)  
#    alpha_high      = Scaling of perturbation by biomass between 0 and 1 (higher bound)  


##############################################################

def observation_simulations(num_funs, num_exp, P_d, P_i, alpha_low, alpha_high):
    
    # vectors to fill in 
    ff_prediction = np.zeros(num_funs)
    ff_prediction_div = np.zeros(num_funs)
    dd_prediction = np.zeros(num_funs)
    bd_prediction = np.zeros(num_funs)
    ff_realised = np.zeros(num_funs)
    dd_realised = np.zeros(num_funs)
    bd_realised = np.zeros(num_funs)
    fb_realised = np.zeros(num_funs)
    bias = np.zeros(num_funs)
    q_diff = np.zeros(num_funs)
    Unevenness = np.zeros(num_funs)
    diversity = np.zeros(num_funs)
    scaling = np.zeros(num_funs)
    ff_Theta = np.zeros(num_funs)
    ff_Theta_pred = np.zeros(num_funs)
    Richness = np.zeros(num_funs)
    function_1_broad = np.zeros(num_funs)

    # Loop to create many different communities and functions
    for n in range(num_funs):
        
        # generate S 
        S = np.random.randint(20, 100)              
        
        Richness[n] = S

        # generate community 
        zeta = np.random.uniform(0.5, 2.0)              # unevenness
        Unevenness[n] = zeta                            # store information
        K = np.exp(np.random.normal(0, zeta, size=S))   # Carrying capacity drawn from log normal distribution 
        Nf = np.maximum(K, 0)                           # Equilibrium biomass
        diversity[n] = hill(Nf, 2)                      # Initial diversity (Hill-Simpson)
        
        # Generate a value of alpha, that controls how much stressor is scaled by K 
        alpha = np.random.uniform(alpha_low, alpha_high)
        scaling[n] = alpha

        ######### 1) Comparing observations of linear function ###########
        
        # generate functions
        narrowness_1 = np.random.uniform(0, 1.3)
        narrowness_2 = np.random.uniform(0, 1.3)
        # positive functions
        function_1 = np.exp(np.random.normal(0, narrowness_1, size=S))            
        function_2 = np.exp(np.random.normal(0, narrowness_2, size=S)) 
        # functions with some negative components (supporting figure)
        #function_1 = np.random.normal(0, narrowness_1, size=S)            
        #function_2 = np.random.normal(0, narrowness_2, size=S)

        # functions scaled by K^alpha
        function_1_mod = (K**alpha) * function_1
        function_2_mod = (K**alpha) * function_2
        
        # broadness of function 1
        p = abs(function_1)/sum(abs(function_1))
        function_1_broad[n] = (1/(sum(p**2)))/S

        # measure angle between functions (in radians)
        ff_cos_theta = np.dot(function_1_mod, function_2_mod) / (norm(function_1_mod) * norm(function_2_mod))
        ff_theta = np.arccos(ff_cos_theta)   
        
        # predict angle between functions based on diversity of functions
        ff_predicted_cos_theta = (np.sqrt(hill(function_1, 2) * hill(function_2, 2)))/S              
        ff_predicted_theta = np.arccos(ff_predicted_cos_theta)

        # measure and store prediction of mismatches based on actual angle between function
        ff_prediction[n] = ff_theta/np.pi   
        
        # measure and store prediction of mismatches based on predicted angle between function
        ff_prediction_div[n] = ff_predicted_theta/np.pi     
        
        # store measuted and predicted angles between function 
        ff_Theta[n] = ff_theta
        ff_Theta_pred[n] = ff_predicted_theta

        
        ######### 2) Comparing observations of diversity metrics ###########
        
        # generate diversity indices and their gradients 
        p = K/np.sum(K)  
        q1 = np.random.uniform(0.01, 5)
        q2 = np.random.uniform(0.01, 5)
        d_grad_1 = (np.power(p, q1-1) - np.sum(np.power(p, q1))) * (q1/(1-q1))
        d_grad_2 = (np.power(p, q2-1) - np.sum(np.power(p, q2))) * (q2/(1-q2))
        q_diff[n] = abs(q1-q2) # store the difference in q values of the two diversity metrics

        # diversity gradients scaled by K^alpha
        d_grad_1_mod = (K**alpha) * d_grad_1
        d_grad_2_mod = (K**alpha) * d_grad_2

        # measure angle between diversity indices gradients (in radians)
        dd_cos_theta = np.dot(d_grad_1_mod, d_grad_2_mod) / (norm(d_grad_1_mod) * norm(d_grad_2_mod)) 
        dd_theta = np.arccos(dd_cos_theta)                                               

        # prediction of mismatches based on angle between diversity indices gradients
        dd_prediction[n] = dd_theta/np.pi 
    
        
        ######### 3) Comparing observations of biomass and diversity ########## 
        
        # generate total biomass and gradient of a diversity index 
        p = K/np.sum(K)  
        total = np.ones(S) 
        q = 2  # Hill-Simpson
        d_grad = (np.power(p, q-1) - np.sum(np.power(p, q))) * (q/(1-q))

        # gradients scaled by K^alpha
        total_mod = (K**alpha) * total
        d_grad_mod = (K**alpha) * d_grad

        # measure angle between biomass and gradient of diversity index 
        bd_cos_theta = np.dot(d_grad_mod, total_mod) / (norm(d_grad_mod) * norm(total_mod))    
        bd_theta = np.arccos(bd_cos_theta)                                                

        # prediction of mismatches based on angle between biomass and gradient of diversity index 
        bd_prediction[n] = bd_theta/np.pi 
        
        ### vectors to fill in during second loop ###
        ff_mismatch = np.zeros(num_exp)
        dd_mismatch = np.zeros(num_exp)
        bd_mismatch = np.zeros(num_exp)
        fb_mismatch = np.zeros(num_exp)
        frac = np.zeros(num_exp)

        # Generate a bias for the perturbation at the population level
        P_bias = np.random.uniform(-P_d, P_d)  

        # Loop to run perturbation experiments for each community 
        for m in range(num_exp): 

            ########## Generate Perturbation ##########

            # Perturbation
            P = np.random.normal(P_bias, 1, size=S)  # Draw perturbation from a random normal distribution with a mean direction that introduces a bias
            P = P/norm(P)                            # Normalise so that changing direction does not change intensity
            P_int = np.random.uniform(0, P_i)        # Intensity value to scale entire perbation vector by 
            P = P * P_int * (K**alpha)               # Pertubation multiplied by intensity and scaled by K^alpha
            P = P * np.sqrt(S)                       # Multiply by square root of S so intensity becomes a fraction of each species' K

            ########## Apply Perturbation ##########       
            NP = np.maximum((K - P), 0)                 # Biomass after perturbation 
            DS = NP - Nf                                # Effect of perturbation (at population level)
            frac[m] = len(np.where(DS<0)[0])/len(DS)    # fraction of positive v negative - 0.5 would be unbiased (measure of consistency/bias of perturbation at population-level)

            ######### 1) Linear function ###########
            D_fun_1 = np.dot(function_1, DS)            # observation of function 1 - i.e. projecting vector of change onto function 1
            D_fun_2 = np.dot(function_2, DS)            # observation of function 2 - i.e. projecting vector of change onto function 2
            
            ########## 2) Diversities #########     
            D_q1 = hill(NP, q1) - hill(Nf, q1)          # observation of hill q1
            D_q2 = hill(NP, q2) - hill(Nf, q2)          # observation of hill q2 

            ########## 3) Diversity - Biomass #########   
            D_q = hill(NP, q) - hill(Nf, q)             # observation of diversity metric
            T_1 = sum(NP) - sum(Nf)                     # observation of total biomass  
 
            ########## Mismatches (function returns 1 if mismatch, 0 if consistent) #########
            ff_mismatch[m] = is_mismatched(D_fun_1, D_fun_2)     
            dd_mismatch[m] = is_mismatched(D_q1, D_q2)
            bd_mismatch[m] = is_mismatched(D_q, T_1)
            
            # also need mismatch between function 1 and biomass 
            fb_mismatch[m] = is_mismatched(D_fun_1, T_1)
            
        # Actual proportion of mismatches
        ff_realised[n] = np.sum(ff_mismatch)/len(ff_mismatch)    
        dd_realised[n] = np.sum(dd_mismatch)/len(dd_mismatch)
        bd_realised[n] = np.sum(bd_mismatch)/len(bd_mismatch)
        fb_realised[n] = np.sum(fb_mismatch)/len(fb_mismatch)

        
        # Actual population-level consistency/bias
        bias[n] = abs(0.5 - np.mean(frac))                 

        # Ticker
        if n%100==0:
            print(n)
    
    print("done")
        
    return Richness, ff_prediction, ff_prediction_div, dd_prediction, bd_prediction, ff_realised, dd_realised, bd_realised, bias, q_diff, Unevenness, diversity, scaling, ff_Theta, ff_Theta_pred, function_1_broad, fb_realised
```

### Supporting Figure reproducing empirical results  

```python
num_exp = 1000 
S = 50
P_d = 0.3
P_i = 0.1
P_bias = 0

# vectors to fill in 
D_biomass = np.zeros(num_exp)
D_diversity = np.zeros(num_exp)
mismatch = np.zeros(num_exp)
bd_prediction = np.zeros(num_exp)

# Loop to run stressor experiments for each community 
for m in range(num_exp): 
        
    # generate scaling
    alpha = np.random.uniform(0.5, 1)
    
    # generate community 
    S = np.random.randint(20, 100)
    zeta = np.random.uniform(0.5, 2.5)      # unevenness
    K = np.exp(np.random.normal(0, zeta, size=S))    
    Nf = np.maximum(K, 0)
    p = K/np.sum(K)

    # generate total biomass 
    total = np.ones(S) 
    total_mod = (K**alpha) * total
    
    # generate diversity 
    q = np.random.uniform(0.01, 2)
    d_grad = (np.power(p, q-1) - np.sum(np.power(p, q))) * (q/(1-q))
    d_grad_mod = (K**alpha) * d_grad

    # generate perturbation
    P = np.random.normal(P_bias, 1, size=S)  # Draw perturbation from a random normal distribution with a mean direction
    P = P/norm(P)                            # Normalise so that changing direction does not change intensity
    P_int = np.random.uniform(0, P_i)        # Intensity value to scale entire perbation vector by 
    P = P * P_int * (K**alpha)               # Pertubation multiplied by intensity and scaled by K^alpha
    P = P * np.sqrt(S)                       # Multiply by square root of S so intensity becomes a fraction of each species' K

    ########## apply perturbation ##########       
    NP = np.maximum((K - P), 0)        # biomass after perturbation
    D_q = hill(NP, q) - hill(Nf, q)    # observation of diversity
    D_diversity[m] = D_q               
    D_T = sum(NP) - sum(Nf)            # observation of total biomass
    D_biomass[m] = D_T                
    
    # Mismatches
    mismatch[m] = is_mismatched(D_T, D_q)  # 1 is mismatch, 0 if same sign 
```
