```python
pd.read_csv("all.csv")     #pooled biomes
#pd.read_csv("forest.csv") #forest biome
#pd.read_csv("grass.csv")  #grassland biome
```

```python
biome='grass' #grass, forest,  all

if biome=='grass':
    data=pd.read_csv("grass.csv")  #grassland biome
if biome=='forest':
    data=pd.read_csv("forest.csv")  #forest biome
if biome=='all':
    data=pd.read_csv("all.csv")  #both biomes


Pij=data.values[:,1:]
K=len(Pij[0:])
print('number of functions=', K)
Cos_data=np.zeros([K,K])
for i in range(K):
    for j in range(K):
        Cos_data[i,j]=np.cos(Pij[i,j]*np.pi)
```

```python
Norm=np.zeros([K,K])
Cos_hat=np.zeros([K,K])

for i in range(K):
    for j in range(K):
        Cij=np.delete(np.delete(Cos_data,[i,j],0),[i,j],1)
        R=inv(Cij)
        Norm[i,j]=norm(R)
        u=np.delete(Cos_data[i,:],[i,j])
        v=np.delete(Cos_data[j,:],[i,j])
        Cos_hat[i,j]=np.dot(np.dot(R,u),v)
        up=np.delete(Pij[i,:],[i,j])
        vp=np.delete(Pij[j,:],[i,j])
        
cos=np.ravel(Cos_data)    #ground truth
cos_hat=np.ravel(Cos_hat) #estimation
normM=np.ravel(Norm)      #predicted sensitivity

roof=15 #upper bound on sensitivity

if biome=='grass':
    plt.scatter(cos_hat[normM<roof],cos[normM<roof],marker='X',color='orange',s=1000/normM[normM<roof])
    plt.title('grasslands soil microbiota', fontsize=14)
if biome=='forest':
    plt.scatter(cos_hat[normM<roof],cos[normM<roof],marker='^',color='green',s=1000/normM[normM<roof])
    plt.title('forests soil microbiota', fontsize=14)
if biome=='all':
    plt.scatter(cos_hat[normM<roof],cos[normM<roof],s=1000/normM[normM<roof])
    plt.title('grasslands and forests soil microbiota (pooled)', fontsize=14)


plt.plot([-1,1],[-1,1],'--k')
plt.vlines(1,-1,1.05)
plt.vlines(-1,-1,1.05)
plt.fill_between([1,2],[-1,-1],[1.05,1.05],color='red',alpha=0.3)
plt.fill_between([-2,-1],[-1,-1],[1.05,1.05],color='red',alpha=0.3)

plt.xlabel('$\hat{C}_{ij}$',fontsize=14)
plt.ylabel('$C_{ij}=\cos(\pi P_{ij})$',fontsize=14)
plt.ylim(-1,1.05)
plt.xlim(-2,2)

#plt.title('grasslands and forests soil microbiota (pooled)', fontsize=14)

#linear fit
x=np.concatenate([cos_hat[normM<roof],cos_hat[normM<roof]])
y=np.concatenate([cos[normM<roof],cos[normM<roof]])

q_tot=np.polyfit(x,y,1)
z=np.linspace(-2,2,100)
#plt.plot(z,z*q_grass[0]+q_grass[1])
#plt.plot(z,z*q_forest[0]+q_forest[1])
plt.plot(z,z*q_tot[0]+q_tot[1],'k-',lw=1)
```

```python
Prod=np.zeros([K-1,K-1])

for i in range(K-1):
    for j in range(K-1):
        Prod[i,j]=Cos_data[i,-1]*Cos_data[-1,j]

Cos_bio=np.delete(np.delete(Cos_data,K-1,0),K-1,1) 
normM_red=np.delete(np.delete(Norm,K-1,0),K-1,1)

#linear fit
x=np.ravel(Prod[normM_red<roof])
y=np.ravel(Cos_bio[normM_red<roof])
p_all=np.polyfit(x,y,1)
z=np.linspace(-1,1,100)
plt.plot(z,z*p_all[0]+p_all[1],'k-',lw=1)


if biome=='grass':
    plt.scatter(Prod[normM_red<roof],Cos_bio[normM_red<roof],marker='X',color='orange',s=50)
    plt.title('grasslands soil microbiota', fontsize=14)
if biome=='forest':
    plt.scatter(Prod[normM_red<roof],Cos_bio[normM_red<roof],marker='^',color='green',s=50)
    plt.title('forests soil microbiota', fontsize=14)
if biome=='all':
    plt.scatter(Prod[normM_red<roof],Cos_bio[normM_red<roof],s=80)
    plt.title('grasslands and forests soil microbiota (pooled)', fontsize=14)

#plt.scatter(Prod[normM_red<roof],Cos_bio[normM_red<roof],s=80)
#plt.scatter(Prod_forest[normM_forest_red<roof],Cos_bio_forest[normM_forest_red<roof],marker='^',color='green',s=50)
#plt.scatter(Prod_grass[normM_grass_red<roof],Cos_bio_grass[normM_grass_red<roof],marker='X',color='orange',s=50)

#plt.title('grasslands and forests (pooled) soil microbiota', fontsize=14)
#plt.title('grasslands (x) and forests ($\Delta$) soil microbiota', fontsize=14)

plt.xlabel('$\cos(\pi P_{i,bio})\cos(\pi P_{j,bio})$',fontsize=14)
plt.ylabel('$\cos(\pi P_{ij})$',fontsize=14)
plt.plot([-1,1],[-1,1],'--k')
plt.ylim(-1,1.05)
plt.xlim(-1,1)
```
