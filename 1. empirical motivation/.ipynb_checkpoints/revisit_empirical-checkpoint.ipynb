```python
######## General use #############
import numpy as np      
import matplotlib.pyplot as plt                                         
import pandas as pd                                                
import scipy.stats as stats 
```

Load data

```python
# Read in the empirical data 
all_data = pd.read_csv("output/all.csv", index_col = 0)
grass = pd.read_csv("output/grass.csv", index_col = 0)
forest = pd.read_csv("output/forest.csv", index_col = 0)
forest_and_grass = pd.read_csv("output/forest_and_grass.csv", index_col = 0)
```

Convert to matrix 

```python
all_matrix = all_data.values
grass_matrix = grass.values
forest_matrix = forest.values
forest_and_grass_matrix = forest_and_grass.values
```

Get cosine matrix

```python
Cos_all = np.cos(np.multiply(all_matrix, np.pi))
Cos_grass = np.cos(np.multiply(grass_matrix, np.pi))
Cos_forest = np.cos(np.multiply(forest_matrix, np.pi))
Cos_forest_and_grass = np.cos(np.multiply(forest_and_grass_matrix, np.pi))
```

Make predictions

1. All

```python
K = np.shape(Cos_all)[0]
Norm = np.empty_like(Cos_all)
Sim = np.empty_like(Cos_all)
Cos = Cos_all

for i in range(K):
    for j in range(K):
        Cij=np.delete(np.delete(Cos,[i,j],0),[i,j],1)
        Mij=np.linalg.inv(Cij)
        Norm[i,j]=np.linalg.norm(Mij)
        u=np.delete(Cos[i,:],[i,j])
        v=np.delete(Cos[j,:],[i,j])
        Sim[i,j]=np.dot(np.dot(Mij,u),v)

cos=np.ravel(Cos)
sim=np.ravel(Sim)
normM=np.ravel(Norm)

maxN = 50

plt.scatter(cos[normM<=maxN], sim[normM<=maxN], c=normM[normM<=maxN])
plt.plot(np.sort(cos[normM<=maxN]), np.sort(cos[normM<=maxN]))
#plt.ylim(-1, 1)
plt.colorbar();
stats.pearsonr(cos[normM<=maxN], sim[normM<=maxN])[0]
```
